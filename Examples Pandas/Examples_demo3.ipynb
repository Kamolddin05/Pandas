```python
df = pd.read_csv("car_prices.csv")
df.head()
```

```python
df["engineV"] = df["engine"].str.extract(r"(\d+\.?\d*)L").astype(float)
df["price_per_engine"] = df["price"] / df["engineV"]
df.pivot_table(index="brand", values= "price", aggfunc= np.mean)
```

```python
df["engineV"] = df["engine"].str.extract(r"(\d+\.?\d*)L").astype(float)
df.groupby(["brand", "model_year", "fuel_type"]).agg(
    avg_price = ("price", "mean"),
    max_engine = ("engineV", "max"),
    median_milage = ("milage", "median"),
    total_cars = ("price", "count")
    
)
```
