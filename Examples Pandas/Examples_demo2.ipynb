```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("car_prices.csv")
df.head()
```

```python
#""" fuel_type ustuni “Petrol” yoki “Hybrid” bo‘lgan mashinalarni tanlab, ularni price bo‘yicha kamayish tartibida tartibla. """#
```

```python
df[df["fuel_type"].isin(["Petrol", "Hybrid"])].sort_values(by = "price", ascending= False)
```

```python
#"""    brand ustunida “BMW” so‘zi qatnashgan mashinalarni topib, model_year bo‘yicha o‘sish tartibida sortla. """#
```

```python
df[df["brand"].isin(["BMW"])].sort_values(by = "model_year", ascending= True)
```

```python
#""" transmission qiymati NaN bo‘lmagan mashinalarni saralab, milage bo‘yicha eng kam yurilgan 10 ta mashinani chiqarsin.  """#
```

```python
df[df["transmission"].notnull()].sort_values("milage").head(10)
```

```python
# ----- price qiymati o‘rtacha narxdan yuqori bo‘lgan mashinalarni tanlab, brand va price bo‘yicha o‘sish tartibida sortla. ----- #
```

```python
c = df["price"].notnull().sum()
print(f"Price NaN value: {c}")

a = df["price"].sum()
b = df["price"].count()

avg = a / b

print(f"Price car Sum: {a}" )
print(f"Price car count: {b}")
print(f"Average value: {avg}")
```

```python
# ----- ext_col ustunida “Black” so‘zi qatnashgan mashinalarni topib, ularni model_year bo‘yicha kamayish tartibida sarala ----- #
```

```python
df[df["ext_col"].isin(["Black"])].sort_values(by = "model_year", ascending= False)
```

```python
# ----- fuel_type ustunida “Diesel” bo‘lgan va engine hajmi 2.0L dan katta mashinalarni tanlab, price bo‘yicha o‘sish tartibida sarala. ----- #
```
