```python
arr1 = np.random.randint(1, 10, (3, 3))
arr2 = np.random.randint(1, 10, (3, 3))
arr3 = np.random.randint(1, 10, (3, 3))
arr_3D = np.array([arr1, arr2, arr3])

for i in range(arr_3D.shape[0]):
    plt.imshow(arr_3D[i], cmap='viridis', interpolation='none')
    plt.title(f"Layer {i+1}")
    plt.colorbar()
    plt.show()
```
